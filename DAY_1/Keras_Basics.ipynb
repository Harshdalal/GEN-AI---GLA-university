#üîÅ What is Sequential in Keras?

‚úÖ Definition:

Sequential is a linear stack of layers in Keras. You use it when your model has one input and one output and layers are stacked one after another, without any branching or merging.

üß† Think of it like:

Input ‚Üí Layer 1 ‚Üí Layer 2 ‚Üí ... ‚Üí Output


```python
#example

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(16, activation='relu', input_shape=(10,)),
    Dense(1, activation='sigmoid')
])
```

image.png

#Use Sequential when:

>You want to build a simple feedforward model

>No multiple inputs/outputs or shared layers

#üß± What is Dense in Keras?

‚úÖ Definition:

Dense is a fully connected layer where:

>Every neuron in the layer is connected to every neuron in the previous layer.

>Each neuron computes a weighted sum + bias, followed by an activation function.

```python
from tensorflow.keras.layers import Dense

Dense(64, activation='relu')
```

#This means:

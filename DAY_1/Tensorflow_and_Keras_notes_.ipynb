#TensorFlow

TensorFlow is an open-source platform by Google for numerical computation and large-scale machine learning.

>Built for deep learning tasks.

>Provides both low-level APIs (e.g., tf.GradientTape) and high-level APIs (e.g., tf.keras).

>Optimized for both CPUs and GPUs.

#Keras

Keras is a high-level neural network API, written in Python and running on top of TensorFlow.

>It simplifies building and training neural networks.

>Integrated into TensorFlow as tf.keras.

#✅ Core Concepts in Keras & TensorFlow
#1. Model Building: Sequential vs Functional API


Sequential API

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Generate sample data for demonstration
X_train = np.random.rand(100, 10)  # 100 samples, 10 features
y_train = np.random.randint(0, 2, 100) # 100 samples, binary labels

# Generate sample data for demonstration
X_test = np.random.rand(20, 10)  # 20 samples, 10 features
y_test = np.random.randint(0, 2, 20) # 20 samples, binary labels

# Generate sample data for prediction
X_new = np.random.rand(5, 10)  # 5 new samples, 10 features

# Simple feed-forward network with input, hidden, and output layer
model = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),  # Hidden layer
    Dense(1, activation='sigmoid')                    # Output layer
])
```

image.png

#2. Compilation

```python
model.compile(optimizer='adam',  # Optimization algorithm
              loss='binary_crossentropy',  # Loss function
              metrics=['accuracy'])        # Evaluation metric
```

#3. Training

```python

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

#epochs: Number of full passes over training data.
#batch_size: How many samples per gradient update.
#validation_split: Hold-out fraction for validation.
```

#4. Evaluation

```python
model.evaluate(X_test, y_test)
```

#5. Prediction

```python
predictions = model.predict(X_new)
```

#🧠 Example: Binary Classification with Keras

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Sample dummy data
import numpy as np
X_train = np.random.rand(100, 10)
y_train = np.random.randint(0, 2, 100)

# Build model
model = Sequential([
    Dense(16, activation='relu', input_shape=(10,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(X_train, y_train, epochs=10, batch_size=8, validation_split=0.2)
```

#🎯 Bonus: TensorFlow Low-Level API Example

```python
# Using tf.GradientTape for manual training
x = tf.Variable(3.0)
with tf.GradientTape() as tape:
    y = x**2
grad = tape.gradient(y, x)
print("dy/dx:", grad.numpy())  # Output: 6.0
```

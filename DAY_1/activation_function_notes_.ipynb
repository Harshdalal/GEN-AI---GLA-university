#Activation Function

An activation function adds non-linearity to neural networks and helps them learn complex patterns.

It transforms the input signal to an output signal ‚Äî often squashing it into a range like [0,1] or [-1,1].

#üîù Top Activation Functions

| Name       | Output Range | Use Case                            |
| ---------- | ------------ | ----------------------------------- |
| Sigmoid    | (0, 1)       | Binary classification (output)      |
| Tanh       | (-1, 1)      | Hidden layers, better than Sigmoid  |
| ReLU       | \[0, ‚àû)      | Hidden layers in deep networks      |
| Leaky ReLU | (-‚àû, ‚àû)      | Like ReLU, but handles "dying ReLU" |
| Softmax    | \[0, 1]      | Multiclass classification (output)  |


# 1. Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

#üìò 2. Tanh (Hyperbolic Tangent)
Equation:

image.png

image.png

```python
def tanh(x):
    return np.tanh(x)
```

#üìò 3. ReLU (Rectified Linear Unit)
Equation:

image.png

image.png

```python
def relu(x):
    return np.maximum(0, x)
```

#üìò 4. Leaky ReLU
Equation:

image.png

image.png

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

#üìò 5. Softmax
Used in output layer of multi-class classification.

Equation:

image.png

image.png

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stability trick
    return exp_x / exp_x.sum(axis=0)
```

#üìä Visual Comparison

```python
x = np.linspace(-10, 10, 100)

plt.figure(figsize=(12, 6))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, leaky_relu(x), label='Leaky ReLU')
plt.plot(x, softmax(x), label='Softmax (normalized)', linestyle='--')

plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

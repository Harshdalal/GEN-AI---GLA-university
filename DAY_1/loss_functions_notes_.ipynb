#Loss Function

A loss function is a mathematical function used to measure the error between the predicted output of a model and the actual ground truth.

>The goal of training a machine learning or deep learning model is to minimize the loss function so that the model predictions become more accurate.

#üîù Top Loss Functions

| Name                      | Use Case                   | Type           |
| ------------------------- | -------------------------- | -------------- |
| Mean Squared Error (MSE)  | Regression                 | Regression     |
| Mean Absolute Error (MAE) | Regression                 | Regression     |
| Binary Cross Entropy      | Binary classification      | Classification |
| Categorical Cross Entropy | Multi-class classification | Classification |
| Hinge Loss                | SVM-style classification   | Classification |

#üî¢ 1. Mean Squared Error (MSE)

Measures the average of squared differences between actual and predicted values.

üìò Equation:

image.png

```python
import numpy as np

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

#üî¢ 2. Root Mean Squared Error (RMSE)

Square root of MSE; provides error in the same units as the target.

üìò Equation:

image.png

```python
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```

#üî¢ 3. Mean Absolute Error (MAE)

üìò Definition: Measures the average of absolute differences between actual and predicted values.

image.png

```python
def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))
```

#üìå Classification Losses:
Used when output is categorical (class labels).

#üî¢ 4. Binary Cross Entropy (Log Loss)

üìò Definition: Measures error for binary classification where output is a probability (0 to 1).

image.png

```python
def binary_cross_entropy(y_true, y_pred, eps=1e-15):
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

#üî¢ 5. Categorical Cross Entropy (CCE)
üìò Definition: Used in multi-class classification (typically with softmax outputs).

image.png

```python
def categorical_cross_entropy(y_true, y_pred, eps=1e-15):
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return -np.sum(y_true * np.log(y_pred)) / y_true.shape[0]
```

#üî¢ 6. Hinge Loss
üìò Definition: Used with Support Vector Machines; penalizes wrong classifications with a margin.

image.png

```python
def hinge_loss(y_true, y_pred):
    return np.mean(np.maximum(0, 1 - y_true * y_pred))

```

#üî¨ Visualizing MSE, MAE, RMSE Comparison

```python
import matplotlib.pyplot as plt

errors = np.linspace(-2, 2, 100)
mse_vals = errors ** 2
mae_vals = np.abs(errors)
rmse_vals = np.sqrt(mse_vals)

plt.plot(errors, mse_vals, label='MSE')
plt.plot(errors, mae_vals, label='MAE')
plt.plot(errors, rmse_vals, label='RMSE', linestyle='--')
plt.title("MSE vs MAE vs RMSE")
plt.xlabel("Error")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

#🧠 Convolutional Neural Network (CNN) – Architecture Explained Step-by-Step
#1. Input Layer
>What it does: Takes image input (e.g., 128×128×3 for RGB images).

>Why it matters: Sets the base dimensions of the data flowing through the network.

#2. Convolutional Layer (Conv Layer)
>Purpose: Extract local features from the image (like edges, textures, patterns).

How:

>Applies multiple filters/kernels (e.g., 32 filters of 3×3).

>Each filter slides (convolves) across the image to produce feature maps.

>Learned Parameters: Filter weights.

✅ Key Concepts: Local connectivity, weight sharing.

#3. Activation Function (ReLU)
>Purpose: Adds non-linearity, allowing the network to learn complex patterns.

>Why ReLU?: Fast, avoids vanishing gradient, sets all negative values to zero.

#4. Pooling Layer (e.g., MaxPooling)

>Purpose: Downsamples the feature map to reduce spatial dimensions (e.g., 128×128 → 64×64).

Type:

>MaxPooling picks the maximum value in a region (e.g., 2×2).

Benefits:

>Reduces computation.

>Adds spatial invariance (important for image position changes).

#5. Repetition of Conv + ReLU + Pool
>Why?: Deeper layers capture higher-level features:

>Layer 1: edges

>Layer 2: textures

>Layer 3: parts of objects (like rashes or lesions)

>Later: complex shapes or disease patterns

#6. Flattening Layer
>Converts 2D feature maps into a 1D vector (needed for Dense layers).

>For example: shape (16, 16, 64) → (1, 16384)

#7. Fully Connected Layer (Dense Layer)
>Purpose: Final classification logic.

>What it does: Takes extracted features and learns weights to map to output labels.

>Usually 1–2 hidden dense layers followed by an output layer.

#8. Output Layer

>For multi-class classification:

Activation: Softmax

Output: Probabilities of each class (e.g., 20 diseases → 20 outputs).

>For binary classification:

Activation: Sigmoid

Output: Single value between 0 and 1.

#9. Loss Function
Measures the error between prediction and true label.

Common: categorical_crossentropy (multi-class), binary_crossentropy (binary).

#10. Optimization (e.g., Adam, SGD)
>Adjusts the weights using backpropagation + gradient descent.

>🔁 Feedback Loop: Backpropagation

>Calculates gradients of loss w.r.t weights (using chain rule).

>Updates all weights from output to input layer.

#11. Training Phase

Feed batches of images.

Model updates its filters and dense layer weights to minimize loss.

#12. Evaluation + Deployment
Test model accuracy on unseen data.

Integrate with MLflow to track performance.

Use Streamlit or web frameworks for user interface.

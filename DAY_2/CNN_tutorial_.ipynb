# Tutorial: Convolutional Neural Network (CNN)

A Convolutional Neural Network (CNN) is a deep learning architecture designed specifically for processing grid-like data, such as images. It automatically learns spatial hierarchies of features, which makes it powerful for tasks like image classification, object detection, and more.

#üß© Key Components of CNN

| **Component**                        | **Definition & Purpose**                                                                                                       |
| ------------------------------------ | ------------------------------------------------------------------------------------------------------------------------------ |
| üîç **Convolutional Layer**           | Applies a set of learnable filters to the input, producing feature maps that capture patterns like edges, textures, or shapes. |
| üì¶ **Kernel/Filter**                 | A small matrix that slides (convolves) over the input to detect patterns.                                                      |
| ‚úÖ **Activation Function (ReLU)**     | Adds non-linearity by zeroing out negative values. Helps model complex patterns.                                               |
| üîΩ **Pooling Layer (MaxPooling)**    | Downsamples the feature maps to reduce computation and capture dominant features.                                              |
| üîÅ **Flatten Layer**                 | Converts the 2D output of convolutions into 1D to feed into dense (fully connected) layers.                                    |
| üîó **Dense (Fully Connected) Layer** | Standard neural network layer where every node is connected to every node in the next layer. Used for classification.          |
| üîÑ **Dropout Layer**                 | Randomly disables a fraction of neurons during training to prevent overfitting.                                                |
| ‚öñÔ∏è **Batch Normalization**           | Normalizes activations between layers to stabilize and speed up training.                                                      |
| üìà **Softmax Layer**                 | Converts output scores into probabilities across multiple classes.                                                             |

#CNN on MNIST Dataset

```python
# Import libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Reshape to add channel dimension (batch, height, width, channel)
X_train = X_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test  = X_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test  = to_categorical(y_test, 10)

# Build the CNN model
model = Sequential()

# Convolutional Layer 1
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(BatchNormalization())   # Normalize activations
model.add(MaxPooling2D(pool_size=(2, 2)))  # Reduce spatial size
model.add(Dropout(0.25))  # Regularization

# Convolutional Layer 2
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Flatten the output
model.add(Flatten())

# Fully connected (Dense) layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Dropout between dense layers

# Output layer
model.add(Dense(10, activation='softmax'))  # 10 classes

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Print model architecture
model.summary()

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=128,
    verbose=2
)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {accuracy:.4f}")

# Plot training history
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

#üìå Key Comments in Code:

Input Layer: Takes the image in (e.g., 28x28 pixels, 1 channel for grayscale).

Conv Layer 1: 32 filters of size 3x3 detect low-level features like edges.

MaxPooling 1: Reduces dimensions, retains important features.

Dropout: Avoids overfitting by randomly turning off neurons.

Conv Layer 2: Learns more abstract patterns (e.g., shapes).

MaxPooling 2 + Dropout

Flatten: Converts 2D data into 1D vector.

Dense: Learns final decision boundary.

Output: Uses Softmax to give class probabilities.

#LSTM and GRU notes

#🌀 1. Why LSTM/GRU Instead of Simple RNN?
1.Simple RNNs struggle with long-term dependencies due to vanishing/exploding gradients during backpropagation.

2.LSTM and GRU solve this by introducing gates that control memory flow, enabling the model to retain or forget information over long sequences.

#🧠 Interpretation

#🧠 Interpretation:

>Update gate ≈ forget + input gate from LSTM.

>Fewer gates = faster training, fewer parameters.

#📊 4. LSTM vs GRU Comparison

| Feature        | LSTM                               | GRU                                          |
| -------------- | ---------------------------------- | -------------------------------------------- |
| Gates          | 3: Forget, Input, Output           | 2: Update, Reset                             |
| Cell State     | Yes                                | No                                           |
| Memory control | More granular                      | Simpler                                      |
| Training Speed | Slower (more complex)              | Faster                                       |
| Performance    | Slightly better for long sequences | Often comparable or better in small datasets |


#✅ Summary

>Both LSTM and GRU help tackle long-term dependencies by regulating memory.

>LSTM is more expressive, GRU is computationally efficient.

>Both are foundational in NLP, time series, audio processing, and more.

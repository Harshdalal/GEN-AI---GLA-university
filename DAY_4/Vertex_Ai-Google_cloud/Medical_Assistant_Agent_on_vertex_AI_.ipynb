# Medical Assistant Agent on vertex AI

✅ Use Case Example:

Build a "Medical AI Assistant" that gives reliable, factual answers on health questions.

🔹 Step 1: Open Vertex AI Studio

🔹 Step 2: Create a New Prompt Template

🔹 Step 3: Add a System Instruction

Click the ⚙️ "Settings" icon in the top-right of the editor.

Then in the “System Instruction” field, write:

>You are a helpful and accurate medical assistant. Always provide fact-based, concise answers using simple language. Avoid giving advice that could be harmful. If you are unsure, clearly say so and suggest the user consult a real doctor.

🔹 Step 4: Try Different Prompts

Now in the Prompt Input area, try these:

✅ Prompt 1:

> What are the symptoms of vitamin D deficiency?

✅ Prompt 2:

> Can I take paracetamol and ibuprofen together?

🔹 Step 5: Observe the Model Behavior

The model should respond factually and safely.

It will warn or redirect when unsure (as per instruction).

✅ You’ll notice how the system instruction affects:

>Tone (professional, calm)

>Detail (more medical, less casual)

>Responsibility (avoids giving direct diagnosis)
